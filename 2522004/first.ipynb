# 投手分析班

## 担当③    やること

Kick-off (全員): 最初の30分で、「コースは9分割にする？」「四球は打数に含めないよね？」というルールだけ合意する。

Phase 1:手書きでいいので「最終的にどういう画面（グラフ）が見たいか」のデザイン案を作る。

Phase 2:Pythonの可視化ライブラリ（SeabornやMatplotlib）のコードを準備しておく。

Phase 3:②の計算結果を、③のコードに流し込んで完成！

Final Phase:実行結果を参考にスライド作成も行う。

### Phase　1
コース別成績の分析画面（ヒートマップ中心）
9分割のコースごとの打撃結果に特化し、打者がどのコースに強いか・弱いかを直感的に把握することを重視したデザイン。
メインエリア	【コース別打率ヒートマップ】（9分割のストライクゾーン）
9つのコースそれぞれに、計算された**打率（BA）や長打率（SLG）**を色分けして表示する。
赤系（高成績）から青系（低成績）へのグラデーションで、打者の得意・不得意なコースが一目瞭然になる。
サイドパネル上	【総合打撃成績カード】	
「四球を打数に含めない」ルールで算出した以下の主要な数値を大きく表示する。
・打率（BA）
・出塁率（OBP）
・OPS（出塁率＋長打率）
サイドパネル下	【コース別打球結果割合グラフ】	
特定のコース（ヒートマップで選択したコースなど）に来た球に対する「ヒット」「四死球」「凡打」「空振り」などの割合を円グラフや積み上げ棒グラフで表示する。
