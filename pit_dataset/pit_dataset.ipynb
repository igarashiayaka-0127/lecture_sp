```python
import pandas as pd
from IPython.display import display

# ãƒ•ã‚¡ã‚¤ãƒ«å
FILE_PIT = 'stats_pit.csv'
OUTPUT_PIT = 'stats_pit_dataset.csv'
# èª­ã¿è¾¼ã¿ã¯æ–‡å­—åŒ–ã‘ãŒè§£æ¶ˆã—ãŸã‚¨ãƒ³ã‚³ãƒ¼ãƒ‡ã‚£ãƒ³ã‚° (cp932) ã§
ENCODING_READ = 'cp932' 
# æ›¸ãå‡ºã—ã¯æ–‡å­—åŒ–ã‘é˜²æ­¢ã®ãŸã‚ BOM ä»˜ã UTF-8 (utf-8-sig) ã§
ENCODING_WRITE = 'utf-8-sig' 

# --- ğŸ¯ çµã‚Šè¾¼ã¿ã®è¨­å®š ---
target_pitcher_names = ['æˆ¸éƒ·ã€€ç¿”å¾', 'æ‘ä¸Šã€€é Œæ¨¹', 'é¢¯', 'å¤§é‡ã€€é›„å¤§', 'æ —æ—ã€€è‰¯å', 'çŸ³å·ã€€é›…è¦', 'ä¼Šè—¤ã€€å¤§æµ·', 'éˆ´æœ¨ã€€ç¿”å¤©', 'å¹³è‰¯ã€€æµ·é¦¬', 'æ¨ªå±±ã€€é™¸äºº', 'å±±å²¡ã€€æ³°è¼”', 'å¤§æ´¥ã€€äº®ä»‹']
target_year = 2025
NAME_COLUMN = 'player_name'    # æŠ•æ‰‹åã®åˆ—åï¼ˆstats_pit.csv ã®æƒ³å®šï¼‰
YEAR_COLUMN = 'season_year'  # å¹´æ¬¡ã®åˆ—åï¼ˆstats_pit.csv ã®æƒ³å®šï¼‰

try:
    # 1. ãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿
    df_pit = pd.read_csv(FILE_PIT, encoding=ENCODING_READ)
    print(f"âœ… {FILE_PIT} ã‚’ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‡ã‚£ãƒ³ã‚° '{ENCODING_READ}' ã§èª­ã¿è¾¼ã¿ã¾ã—ãŸ (ä»¶æ•°: {len(df_pit)})")
    
except Exception as e:
    print(f"âŒ ã‚¨ãƒ©ãƒ¼: {FILE_PIT} ã®èª­ã¿è¾¼ã¿ã«å¤±æ•—ã—ã¾ã—ãŸã€‚è©³ç´°: {e}")
    raise

# --- ãƒ‡ãƒ¼ã‚¿ã®çµã‚Šè¾¼ã¿ ---
df_filtered = df_pit.copy()

# æŠ•æ‰‹åã®çµã‚Šè¾¼ã¿ (player_nameåˆ—ã‚’ä½¿ç”¨)
if target_pitcher_names and NAME_COLUMN in df_filtered.columns:
    df_filtered = df_filtered[df_filtered[NAME_COLUMN].isin(target_pitcher_names)].copy()
    print(f"ğŸ‘¤ æŠ•æ‰‹: {len(target_pitcher_names)} åã«é™å®šã—ã¾ã—ãŸ (ä»¶æ•°: {len(df_filtered)})")

# å¹´æ¬¡ã®çµã‚Šè¾¼ã¿ (season_yearåˆ—ã‚’ä½¿ç”¨)
if YEAR_COLUMN in df_filtered.columns:
    df_filtered = df_filtered[df_filtered[YEAR_COLUMN] == target_year].copy()
    print(f"ğŸ“… å¹´æ¬¡: {target_year} ã®ãƒ‡ãƒ¼ã‚¿ã«çµã‚Šè¾¼ã¿ã¾ã—ãŸ (ä»¶æ•°: {len(df_filtered)})")
else:
    print(f"âš ï¸ æ³¨æ„: å¹´æ¬¡çµã‚Šè¾¼ã¿ã«ä½¿ç”¨ã™ã‚‹åˆ— ('{YEAR_COLUMN}') ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã§ã—ãŸã€‚çµã‚Šè¾¼ã¿ã¯ã‚¹ã‚­ãƒƒãƒ—ã•ã‚Œã¾ã—ãŸã€‚")


# --- âœ‚ï¸ å¿…è¦ãªåˆ—ã®æŠ½å‡º ---
# ã‚·ãƒ¼ã‚ºãƒ³æˆç¸¾ã«å¿…è¦ãªåˆ—ã‚’é¸æŠï¼ˆä¾‹: ç™»æ¿æ•°(g)ã€å‹åˆ©(w)ã€é˜²å¾¡ç‡(era)ãªã©ï¼‰
selected_columns = [
    YEAR_COLUMN, 'player_id', NAME_COLUMN, 
    'team_name', 'era', 'g', 'gs', 'w', 'l', 'sv', 'hld', 'ip', 'er', 'so', 
    'whip', 'avg' # ERA, WHIP, å¯¾æˆ¦æ‰“ç‡ãªã©ã‚‚å«ã‚ã‚‹
]

existing_columns = [col for col in selected_columns if col in df_filtered.columns]
df_dataset_pit = df_filtered[existing_columns]


# --- ãƒ•ã‚¡ã‚¤ãƒ«ä¿å­˜ï¼ˆæ›¸ãå‡ºã—æ™‚ã«ã‚‚æ­£ã—ã„ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‡ã‚£ãƒ³ã‚°ã‚’æŒ‡å®šï¼‰ ---
try:
    df_dataset_pit.to_csv(OUTPUT_PIT, index=False, encoding=ENCODING_WRITE) 
    print(f"\nâœ… åˆ†æç”¨ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆ `{OUTPUT_PIT}` ã‚’ãƒ•ã‚¡ã‚¤ãƒ«ã¨ã—ã¦ä¿å­˜ã—ã¾ã—ãŸã€‚")
    print(f"âœ… ä¿å­˜ã—ãŸãƒ•ã‚¡ã‚¤ãƒ«ã‚‚æ–‡å­—åŒ–ã‘ã—ã¦ã„ãªã„ã“ã¨ã‚’ç¢ºèªã—ã¦ãã ã•ã„ã€‚")
except Exception as e:
    print(f"\nâŒ ãƒ•ã‚¡ã‚¤ãƒ«ä¿å­˜ä¸­ã«ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸã€‚è©³ç´°: {e}")


print(f"ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®ã‚µã‚¤ã‚º: {len(df_dataset_pit)} ä»¶, åˆ—æ•°: {len(df_dataset_pit.columns)} åˆ—")
# print("--- æœ€çµ‚ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®æœ€åˆã®5è¡Œ ---")
# display(df_dataset_pit.head())
```
